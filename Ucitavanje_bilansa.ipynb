**Okvir podataka** (eng. DataFrame) predstavlja pravougaonu tabelu podataka i sadrži uređenu kolekciju kolona, od kojih svaka može biti različitog tipa vrednosti (numerička, string, logička, itd.).

Ključne karakteristike okvira podataka:

1. **Dvostruko indeksiranje** (reda i kolona):  
    - Svaka kolona ima svoj naziv (label), koji funkcioniše kao ključ u rječniku.  
    - Svaki red ima svoj indeks (index), koji omogućava identifikaciju i pristup.
    
2. **Heterogenost podataka**:  
    - Kolone mogu sadržavati različite tipove podataka (npr. numerički, stringovi, datumi), ali svi podaci unutar jedne kolone moraju biti istog tipa.
    
3. **Struktura slična tabeli**:  
    - DataFrame je poput tabela u bazama podataka ili proračunskih tablica (npr. Excel), ali s dodatnim mogućnostima za manipulaciju podacima.
    
4. **Etičke dimenzije** (labele):  
    - Kolone i redovi imaju nazive (etikete), što omogućava pristup i manipulaciju podacima po imenu umjesto samo indeksima.
    
5. **Podrška za operacije**:  
    - DataFrame omogućava filtriranje, grupisanje, pivotiranje, statističku analizu, spajanje i mnoge druge napredne operacije.

Jedna od ključnih karakteristika NumPy-a je njegov N-dimenzionalni objekt niza, ili ndarray, koji je brz, fleksibilan kontejner za velike skupove podataka u Pythonu. Nizovi vam omogućavaju daizvoditi matematičke operacije na cijelim blokovima podataka koristeći sličnu sintaksu kao ekvivalentne operacije između skalarnih elemenata.

# MATPLOTLIB

Matplotlib je Python paket za 2D crtanje koji generiše grafove producijskog kvaliteta. Podržava interaktivno i neinteraktivno crtanje i može pohraniti slike nekoliko izlaznih formata (PNG, PS i drugi). Pruža široku lepezu tipova grafika (linije, trake, tortni grafikoni, histogrami i još mnogo toga). Pored toga, to je visoko prilagodljiv, fleksibilan i jednostavan za korištenje.  

Matplotlib se oslanja na dva glavna objekta: \
1.	**Figure**: Predstavlja celokupni prozor ili sliku koja sadrži jedan ili više grafova. \
2.	**Ose (eng. Axes)**: Predstavlja deo figure gde se podaci zapravo prikazuju. Svaka osa može sadržavati različite vrste      grafova.


Jedna od jačih  strana Matplotlib-a je koliko brzo možemo početi proizvoditi parcele odmah. Evo našeg prvog primjera:

```python
{
  "slide_type": "skip"
}
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 3, 2, 4])
plt.show()
```

Naravno, možemo eksplicitno specificirati Y i X osu.

```python
x = range(6)  #Generiše niz brojeva od 0 do 5 za x-osi.
plt.plot(x, [xi**2 for xi in x]) #Generiše niz vrijednosti za y-osi kao kvadrate elemenata iz x.
plt.show() #Prikazuje grafikon na ekranu.
```

Naravno, u Matplotlib-u  postoje mnogo vrsta grafikona, a ne samo linijske prirode. U ovoj bilioteci postoji i komanda za prikaz podataka u legendi.

```python
kategorije = ['K1', 'K2', 'K3']
v1, v2 = [5, 8, 6], [7, 6, 9]
x = np.arange(len(kategorije))

plt.bar(x - 0.2, v1, width=0.4, label='Grupa 1')
plt.bar(x + 0.2, v2, width=0.4, label='Grupa 2')
plt.xticks(x, kategorije)
plt.legend(); plt.show()
```

```python
labels = ['Jabuke', 'Banane', 'Trešnje', 'Grožđe']
vrijednosti = [15, 30, 45, 10]

# Kreiranje pie grafikona
plt.pie(vrijednosti, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Grafikon')
plt.show()
```

## FINANSIJSKA RAVNOTEŽA

Kratkoročna finansijska ravnoteža se utvrđuje odnosom likvidnih i kratkoročnih vezanih sredstava, s jedne strane i kratkoročnih obaveza sa druge strane.
Da bi se postigla kratkoročna finansijska ravnoteža, koeficijent bi trebao biti veći ili jednak nuli.

```python
Kratkorocna_potrazivanja = bs.loc[45, [2021, 2022, 2023]]
Kratkoročni_finansijski_plasmani = bs.loc [52, [2021, 2022, 2023]]                    \
                                   .fillna(0)
Gotovina_i_gotovinski_ekvivalenti = bs.loc [61, [2021, 2022, 2023]]
Porez_na_dodatu_vrijednost = bs.loc [64, [2021, 2022, 2023]]
Aktivna_kratkoročna_razgraničenja = bs.loc [65, [2021, 2022, 2023]].fillna(0)
kratkoročne_obaveze = bs.loc[114, [2021 ,2022, 2023]]
KRATKOROČNA_VEZANA_SREDSTVA = Kratkorocna_potrazivanja + \
                        Kratkoročni_finansijski_plasmani + \
                        Gotovina_i_gotovinski_ekvivalenti + \
                        Porez_na_dodatu_vrijednost + \
                        Aktivna_kratkoročna_razgraničenja

KRATKO_FR = KRATKOROČNA_VEZANA_SREDSTVA / kratkoročne_obaveze
KRATKO_FR = KRATKO_FR.apply(lambda x: round(x, 3))

KRATKOROČNA_VEZANA_SREDSTVA_df = pd.DataFrame({
  'Kratkorocna potrazivanja': Kratkorocna_potrazivanja,
  'Kratkoročni finansijski plasmani': Kratkoročni_finansijski_plasmani,
  'Gotovina i gotovinski ekvivalenti': Gotovina_i_gotovinski_ekvivalenti,
  'Porez na dodatu vrijednost': Porez_na_dodatu_vrijednost,
  'Aktivna kratkoročna razgraničenja': Aktivna_kratkoročna_razgraničenja,
  'I Kratkoročna vezana sredstva': KRATKOROČNA_VEZANA_SREDSTVA,
  'II Kratkoročni izvori finansiranja' : kratkoročne_obaveze,
  'KRATKOROČNA FINANSIJSKA RAVNOTEŽA (I/II)': KRATKO_FR
}).T

KRATKOROČNA_VEZANA_SREDSTVA_df
```

## POKAZATELJI LIKVIDNOSTI

Likvidnost preduzeća se definiše kao sposobnost preduzeća da raspoloživim likvidnom imovinom uspjeva permametno izmirivati dospjele kratkoročne obaveze. Ovdje smo koristili tri tipa likvidnosti.

Kod **tekuće likvidnosti**, ne važi „acid test“ 1:1, jer se  zalihe finansiraju iz dugoročnih dugova i kapitala. One se uzimaju u obzir jer se potencijalno mogu konvertovati u kratkoročne oblike likvidne imovine prodajom gotovih proizvoda (roba), brzim trošenjem zaliha, materijala, dovršavanjem nedovršene proizvodnje.

```python
obrtna_sredstva = bs.loc[36, [2021 ,2022, 2023]]
kratkoročne_obaveze = bs.loc[114, [2021 ,2022, 2023]]
Tekuća_likvidnost = obrtna_sredstva / kratkoročne_obaveze

KTL = pd.DataFrame ({
    'Obrtna sredstva' : obrtna_sredstva,
    'Kratkoročne obaveze' : kratkoročne_obaveze,
    'Tekuća likvidnost' : Tekuća_likvidnost
}).T

KTL
```

# KOEFICIJENT OBRTA 


**KOEFICIJENTI OBRTA** su važni finansijski pokazatelji koji se koriste za procjenu efikasnosti korišćenja imovine i kapitala u poslovanju. Obično se koriste da bi se odredilo koliko brzo preduzeće koristi svoju imovinu za generisanje prihoda. Iskazuju koliko se puta obrtna imovina “obrnula” u period za koji se mjeri obrt, a broj dana trajanja obrta iskazuje koliko je dana trajao jedan obrt. 

Što je koeficijent obrta viši (a broj dana obrta manji) efikasnost imovine je veća i veća.

### STRUKTURA POSLOVNIH PRIHODA

**Poslovni prihodi** čine prihodi od prodaje +  povećanje vrijednosti zaliha učinaka – smanjenje vrijednosti zaliha učinaka.

Ako poslovni prihodi sadrže povećanje vrijednosti zaliha učinaka to znači da je proizvodnja bila veća od prodaje i obrnuto, ako poslovni prihodi sadrže smanjenje vrijednosti zaliha učinaka to znači da je prodaja bila veća od proizvodnje.

U strukturi poslovnih prihoda treba da dominiraju prihodi od prodaje jer to je zapravo prihod koji pretvara zalihu proizvoda, roba i usluge u gotovinu bez koje preduzeće ne može da plaća obaveze, a zbog toga ne može ni da opstane.

# RIZIK OSTVARENJA FINANSIJSKOG REZULTATA

```python
pd.set_option('future.no_silent_downcasting', True)
```

U analizi rizika ostvarenja finansijskog rezultata ne uključuju se revalorizacioni dobici i gubici, neposlovni i vanredni prihodi i rashodi zbog toga što su oni privremeni i povremeni i kao takvi ne mogu biti dugoročan osnov ostvarenja finansijskog rezultata. Drugim riječima, u ovu analizu se uključuju poslovni prihodi i rashodi i finansijski prihodi i rashodi tj. finansijski rezultat iz redovnog poslovanja. Rashodi se rasčlanjuju na varijabilne i fiksne.

```python
bu.loc[:, [2021, 2022, 2023]] = \
bu.loc[:, [2021, 2022, 2023]].apply(pd.to_numeric, errors='coerce')
bu.fillna(0, inplace=True)

# Izračunavanja
poslovni_prihodi_2021 = (bu.loc[3, 2021] + bu.loc[9, 2021])
poslovni_prihodi_2022 = (bu.loc[3, 2022] + \
                         bu.loc[4, 2022] + \
                         bu.loc[9, 2022])

poslovni_prihodi_2023 = (bu.loc[3, 2023] + bu.loc[4, 2023])
poslovni_prihodi = [poslovni_prihodi_2021, 
                    poslovni_prihodi_2022, 
                    poslovni_prihodi_2023]

Varijabilni_troskovi = (
    bu.loc[13, [2021, 2022, 2023]] 
   + bu.loc[18, [2021, 2022, 2023]] 
    + bu.loc[25, [2021, 2022, 2023]] 
    + bu.loc[26, [2021, 2022, 2023]]
)
Marža_pokrića = poslovni_prihodi - Varijabilni_troskovi

Fiksni_i_pretFiksni_rashodi = (
bu.loc[12, [2021, 2022, 2023]] + 
bu.loc[14, [2021, 2022, 2023]] + 
bu.loc[17, [2021, 2022, 2023]] )

Rashodi_kamata = bu.loc[36, [2021, 2022, 2023]].fillna(0)

Poslovni_rezultat = \
Marža_pokrića - Fiksni_i_pretFiksni_rashodi

Bruto_finansijski_rezultat = \
Poslovni_rezultat - Rashodi_kamata

# Kreiranje DataFrame-a za prikaz rezultata
Pozicije = [
 poslovni_prihodi, Varijabilni_troskovi, Marža_pokrića, 
 Fiksni_i_pretFiksni_rashodi, Rashodi_kamata, 
 Poslovni_rezultat, Bruto_finansijski_rezultat
]
Pozicije = pd.DataFrame(Pozicije)
Pozicije.columns = [2021, 2022, 2023]

pozicije_nazivi = [
'Poslovni prihodi', 'Varijabilni troškovi', 'Marža pokrića',
'Fiksni i pretežno fiksni rashodi', 'Rashodi kamata', 
'Poslovni rezultat', 'Bruto finansijski rezultat'
]

Pozicije.insert(0, 'Pozicija', pozicije_nazivi)

Pozicije
```

Saglasno prethodnim rečenom, podaci posmatranog preduzeća, predviđeni  za analizu rizika ostvarenja finansijskog rezultata i donje tačke rentabilnosti, su:
- Marža pokrića treba da pokrije varijabilne i fiksne rashode.
- Poslovni finansijski rezultat (poslovni dobitak) predstavlja razliku između marže pokrića i fiksinih i pretežno fiksnih rashoda.
- Bruto dobit predstavlja razliku između poslovnog dobitka i neto rashoda finansiranja.
- Neto dobit predstavlja razlikus između bruto dobiti i poreza i doprinosa iz bruto dobiti.

## ZETA i Z ANALIZA

Analiza kreditne sposobnosti često se kombinuje sa različitim tehnikama i modelima predviđanja lošeg poslovanja firmi. Dva najpoznatija modela za predviđanje lošeg poslovanja firmi su: Altmanova Zeta analiza i Cheeser-ov  ''credit scoring system''.

Edward i Altman je koristio višestruku diskriminacionu analizu za predviđanja bankrotstva firmi, koristeći pri tome različite finansijske pokazatelje. On je od 22 finansijska pokazatelja odabrao 5 koji su po njemu najbolji za predviđanje bankrotstva. Na bazi tih 5 pokazatelja formulisao je tkz. Z model.

```python
Kapital = bs.loc[68, [2021, 2022, 2023]]

Rezervisanja_i_odloženiPoreskiPrihodi =                                                         \
(bs.loc[100, [2021, 2022, 2023]].fillna(0) +                                                   \
 bs.loc[112, [2021, 2022, 2023]].fillna(0) +                                                     \
 bs.loc[113, [2021, 2022, 2023]].fillna(0)).infer_objects(copy=False)  
\
dugoročne_obaveze = bs. loc[104, [2021, 2022, 2023]].fillna(0)

TRAJNI_I_DUGOROČ_KAP = Kapital +                                                                \
Rezervisanja_i_odloženiPoreskiPrihodi                                                           \
+ dugoročne_obaveze                                                                            \

Stalna_imovina = bs.loc[1, [2021, 2022, 2023]]                                                   

NETO_OBRKAP_copy = TRAJNI_I_DUGOROČ_KAP - Stalna_imovina
#NETO_OBRTNI_KAPITAL_copy = pd.DataFrame (NETO_OBRTNI_KAPITAL_copy)
Ukupna_aktiva =  bs.loc [66, [2021, 2022, 2023]]
X1 = NETO_OBRKAP_copy / Ukupna_aktiva

Zadržana_dobit=bs.loc[91, [2021 ,2022, 2023]]
Ukupna_aktiva =  bs.loc [66, [2021, 2022, 2023]]
X2 = Zadržana_dobit / Ukupna_aktiva

EBIT = bu.loc[80, [2021,2022,2023]]
Ukupna_aktiva
X3 = EBIT / Ukupna_aktiva

Sopstveni_kapital = bs.loc[68, [2021, 2022,2023]]
Ukupne_obaveze = bs.loc[104, [2021, 2022, 2023]].fillna(0)                                              \
+ bs.loc[114, [2021, 2022, 2023]]
X4 = Sopstveni_kapital / Ukupne_obaveze

Prihodi_od_prodaje = (bu.loc[3, [2021, 2022, 2023]].fillna(0)                                             \
             + bu.loc[4, [2021, 2022, 2023]].fillna(0))
Ukupna_aktiva
X5 = Prihodi_od_prodaje / Ukupna_aktiva
```

```python
NETO_OBRKAP_copy = NETO_OBRKAP_copy.squeeze()
```

```python
df1 = pd.DataFrame({
    'I Neto obrtni kapital' : NETO_OBRKAP_copy,
    'II Ukupna aktiva' : Ukupna_aktiva,
    'X1 (I/II)' : X1,
    'III Zadržana dobit': Zadržana_dobit,
    'IV Ukupna aktiva' : Ukupna_aktiva,
    'X2 (III/IV)' : X2,
    'V Dobit prije oporezivanja(EBIT)' : EBIT,
    'VI Ukupna aktiva' : Ukupna_aktiva,
    'X3 (V/VI) ' :  X3,
    'VII Tržišna vrijednost (Sopstveni kapital)':
                             Sopstveni_kapital,
    'VIII Ukupne obaveze' : Ukupne_obaveze,
    'X4 (VII / VIII) ' :  X4,
    'IX Prihodi od prodaje': Prihodi_od_prodaje,
    'X Ukupna aktiva': Ukupna_aktiva,
    'X5 (IX/X)': X5
}).T

df1
```

2. Ako je između 1,24 i 2,89 preduzeće ima minimalne kreditne performanse i njegovo poslovanje se ocjenjuje kao rizično. (SIVA ZONA)
