**Okvir podataka** (eng. DataFrame) predstavlja pravougaonu tabelu podataka i sadrži uređenu kolekciju kolona, od kojih svaka može biti različitog tipa vrednosti (numerička, string, logička, itd.).

Naravno, u Matplotlib-u  postoje mnogo vrsta grafikona, a ne samo linijske prirode. U ovoj bilioteci postoji i komanda za prikaz podataka u legendi.

```python
kategorije = ['K1', 'K2', 'K3']
v1, v2 = [5, 8, 6], [7, 6, 9]
x = np.arange(len(kategorije))

plt.bar(x - 0.2, v1, width=0.4, label='Grupa 1')
plt.bar(x + 0.2, v2, width=0.4, label='Grupa 2')
plt.xticks(x, kategorije)
plt.legend(); plt.show()
```

## FUNKCIJE ZA "ČITANJE" POZICIJA U BILANSIMA

```python
def vrednosti_iz_excela(putanja_do_fajla, aop_vrednosti_po_godinama):
    df = pd.read_excel(putanja_do_fajla)
    
    df['AOP'] = df['AOP'].astype(str).str.zfill(3)
    
    rezultati = []
    
    for godina, aop_vrednost in aop_vrednosti_po_godinama.items():
        filtrirani_df = df[(df['AOP'] == aop_vrednost) & (df['Godina'] == godina)]
        
        if not filtrirani_df.empty:
            rezultati.append(filtrirani_df['Neto tekuća'].values[0].item())
        else:
            rezultati.append(None)  
    
    return pd.Series(rezultati)
    # izbriši donju liniju a odkomentariši gornju ako hoces čiste podatke bez godina
    #return pd.Series(rezultati, index=list(aop_vrednosti_po_godinama.keys())) 



# ------------------Primjer dodjele vrednosti varijabli--------------------
# Promeni vrednosti AOP za razlicite god (039,044,044)

Kratkorocna_potrazivanja = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '040', 2022: '045', 2023: '045'}) 
```

```python
def vrednost_iz_excela(putanja_do_fajla, aop_vrednost, godina_vrednost):
    df = pd.read_excel(putanja_do_fajla)

    df['AOP'] = df['AOP'].astype(str).str.zfill(3)  

    
    filtrirani_df = df[(df['AOP'] == aop_vrednost) & (df['Godina'] == godina_vrednost)]
    
    if not filtrirani_df.empty:
        return filtrirani_df['Neto tekuća'].values[0]
    else:
        return None  
    
# ------------------Primjer dodjele vrednosti varijabli--------------------
```

# SOLVENTNOST I ZADUŽENOST

Solventnost označava spospobnost preduzeća da plati sve obaveze (dugove) ne o roku njihovog dospjeća za plaćanje, već kad tad, makar iz stečajne odnosno likvidne mase.

$$
\text{Solventnost} = \frac{\text{Poslovna imovina}}{\text{Dugovi}} 
$$


```python
poslovna_imovina = \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '064', 2022: '066', 2023: '066'}) 
dugovi =  \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '135', 2022: '147', 2023: '147'}) 
solventnost = poslovna_imovina / dugovi
solventnost = pd.DataFrame (solventnost)
solventnost.columns = ['Koeficijent solventnosti']
solventnost.index = ['2021', '2022', '2023']
solventnost = solventnost.apply (lambda x: round (x,3))
solventnost
```
