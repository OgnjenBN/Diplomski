## FUNKCIJE ZA "ČITANJE" POZICIJA U BILANSIMA

```python
def vrednosti_iz_excela(putanja_do_fajla, aop_vrednosti_po_godinama):
    df = pd.read_excel(putanja_do_fajla)
    
    df['AOP'] = df['AOP'].astype(str).str.zfill(3)
    
    rezultati = []
    
    for godina, aop_vrednost in aop_vrednosti_po_godinama.items():
        filtrirani_df = df[(df['AOP'] == aop_vrednost) & (df['Godina'] == godina)]
        
        if not filtrirani_df.empty:
            rezultati.append(filtrirani_df['Neto tekuća'].values[0].item())
        else:
            rezultati.append(None)  
    
    return pd.Series(rezultati)
    # izbriši donju liniju a odkomentariši gornju ako hoces čiste podatke bez godina
    #return pd.Series(rezultati, index=list(aop_vrednosti_po_godinama.keys())) 



# ------------------Primjer dodjele vrednosti varijabli--------------------
# Promeni vrednosti AOP za razlicite god (039,044,044)

Kratkorocna_potrazivanja = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '040', 2022: '045', 2023: '045'}) 
```

```python
def vrednost_iz_excela(putanja_do_fajla, aop_vrednost, godina_vrednost):
    df = pd.read_excel(putanja_do_fajla)

    df['AOP'] = df['AOP'].astype(str).str.zfill(3)  

    
    filtrirani_df = df[(df['AOP'] == aop_vrednost) & (df['Godina'] == godina_vrednost)]
    
    if not filtrirani_df.empty:
        return filtrirani_df['Neto tekuća'].values[0]
    else:
        return None  
    
# ------------------Primjer dodjele vrednosti varijabli--------------------
```

# LIKVIDNOST

Likvidnost preduzeća se definiše kao sposobnost preduzeća da raspoloživom likvidnom imovinom uspjeva permametno izmirivati dospjele kratkoročne obaveze. 

```python
obrtna_sredstva =                                                                                               \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '031', 2022: '036', 2023: '036'}) 

Kratk_obaveze =                                                                                               \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '144', 2022: '147', 2023: '147'}) 
Tekuća_likvidnost = obrtna_sredstva / Kratk_obaveze

KTL = pd.DataFrame ({
    'Obrtna sredstva' : obrtna_sredstva,
    'Kratkoročne obaveze' : Kratk_obaveze,
    'Tekuća likvidnost' : Tekuća_likvidnost
}).T

KTL = KTL.map(
    lambda x: f"{int(x):,}" if isinstance(x, (int, float)) 
    and x == int(x) else f"{x:.3f}" if isinstance(x, float) else x)
KTL.columns = [2021, 2022, 2023]
KTL
```

# SOLVENTNOST I ZADUŽENOST

Solventnost označava spospobnost preduzeća da plati sve obaveze (dugove) ne o roku njihovog dospjeća za plaćanje, već kad tad, makar iz stečajne odnosno likvidne mase.

$$
\text{Solventnost} = \frac{\text{Poslovna imovina}}{\text{Dugovi}} 
$$


```python
poslovna_imovina = \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '064', 2022: '066', 2023: '066'}) 
dugovi =  \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '135', 2022: '147', 2023: '147'}) 
solventnost = poslovna_imovina / dugovi
solventnost = pd.DataFrame (solventnost)
solventnost.columns = ['Koeficijent solventnosti']
solventnost.index = ['2021', '2022', '2023']
solventnost = solventnost.apply (lambda x: round (x,3))
solventnost
```

```python
years = [2021, 2022, 2023]

solventnost = [11.264695, 17.386842, 13.697534]
referentne_vrijednosti = [1, 1, 1]

plt.figure(figsize=(8, 6))
sns.lineplot(x=years, y=solventnost, marker='o', \
             color='orange', label='Solventnost')
sns.lineplot(x=years, y=referentne_vrijednosti, marker='o', \
             color='g', label='Referentne vrijednosti')

plt.xticks(years)  # Prikaz samo godina na x-osi
plt.yticks(range(0, 20, 2))   # Postavljanje y-osi do 13

plt.xlabel('Godine')
plt.ylabel('Vrijednosti')
plt.title('Koeficijent solventnosti ')

plt.legend(); plt.show()
```

## ZETA ANALIZA

Edward Altman je koristio višestruku diskriminacionu analizu za predviđanja bankrotstva firmi, koristeći pri tome različite finansijske pokazatelje. On je od 22 finansijska pokazatelja odabrao 5 koji su po njemu najbolji za predviđanje bankrotstva. Na bazi tih 5 pokazatelja formulisao je tkz. Z model, koji je sa određenom vjerovatnoćom mogao predvidjeti stečaj preduzeća za vremenski period od 1-5 godina.

```python
Kapital = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '101', 2022: '101', 2023: '101'})
Rezervisanja_odlož = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '126', 2022: '132', 2023: '132'}) 
dugoročne_obaveze = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx', {2021: '136', 2022: '137', 2023: '137'}) 
TRAJNI_DUG_KAP = Kapital + Rezervisanja_odlož + dugoročne_obaveze
Ukupna_aktiva = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '064', 2022: '066', 2023: '066'})
Stalna_imovina = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '001', 2022: '001', 2023: '001'})

NETO_OBRKAP_copy = TRAJNI_DUG_KAP - Stalna_imovina
Ukupna_aktiva
X1 = NETO_OBRKAP_copy / Ukupna_aktiva

Zadržana_dobit = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '119', 2022: '124', 2023: '124'})
Ukupna_aktiva
X2 = Zadržana_dobit / Ukupna_aktiva

EBIT = vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '119', 2022: '124', 2023: '124'})
Ukupna_aktiva
X3 = EBIT / Ukupna_aktiva

Sops_kapital = \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '101', 2022: '101', 2023: '101'})
Ukupne_obaveze = \
vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '136', 2022: '137', 2023: '137'}) \
+ vrednosti_iz_excela('BMNT_bilans_stanja.xlsx',{2021: '144', 2022: '147', 2023: '147'})
X4 = Sops_kapital / Ukupne_obaveze

Prihodi_od_prodaje = Prih_učinci + Prih_usluge 
Ukupna_aktiva
X5 = Prihodi_od_prodaje / Ukupna_aktiva

```

```python
df1 = pd.DataFrame({
    'I Neto obrtni kapital' : NETO_OBRKAP_copy,
    'II Ukupna aktiva' : Ukupna_aktiva,
    'X1 (I/II)' : X1,
    'III Zadržana dobit': Zadržana_dobit,
    'IV Ukupna aktiva' : Ukupna_aktiva,
    'X2 (III/IV)' : X2,
    'V Dobit prije oporezivanja(EBIT)' : EBIT,
    'VI Ukupna aktiva' : Ukupna_aktiva,
    'X3 (V/VI) ' :  X3,
    'VII Tržišna vrijednost (Sopstveni kapital)':Sops_kapital,
    'VIII Ukupne obaveze' : Ukupne_obaveze,
    'X4 (VII / VIII) ' :  X4,
    'IX Prihodi od prodaje': Prihodi_od_prodaje,
    'X Ukupna aktiva': Ukupna_aktiva,
    'X5 (IX/X)': X5
}).T
df1_formatted = df1.map(lambda x: f"{x:,.3f}" if isinstance(x, (int, float)) else x)
df1_formatted.columns = [2021, 2022, 2023]
df1_formatted
```

<h2>Kriterijum Z-skora</h2>

<ul>
    <li><strong>Z-skor ≥ 2,90:</strong> Preduzeće ima dobre kreditne performanse i njegovo poslovanje se ocjenjuje kao zdravo, odnosno postoji jaka platežna sposobnost i zarađivačka moć. Takođe, preduzeće se nalazi u <span style="color: green;"><strong>bijeloj zoni</strong></span>.</li>
    <li><strong>1,24 ≤ Z-skor ≤ 2,89:</strong> Preduzeće ima minimalne kreditne performanse i njegovo poslovanje se ocjenjuje kao rizično. Preduzeće se nalazi u <span style="color: gray;"><strong>sivoj zoni</strong></span>.</li>
    <li><strong>Z-skor ≤ 1,23:</strong> Preduzeće nema kreditne performanse, odnosno njegovo poslovanje je pred bankrotstvom. To pokazuje investitorima da ne ulažu u dato preduzeće jer ono ne posjeduje željene performanse i prisutan je visok rizik ulaganja. Preduzeće se nalazi u <span style="color: red;"><strong>crvenoj zoni</strong></span>.</li>
</ul>

```python
Z_analiza =                                                                  \
((0.717*X1)+(0.847*X2)+(3.107*X3)+(0.420 * X4)+(0.948*X5))

Zeta_analiza =                                                               \
((0.00717*X1)+(0.00847*X2)+(0.03107*X3)+(0.00420 * X4)+(0.00948*X5))

Z_i_zeta_analiza = pd.DataFrame ({
    'Z analiza' : Z_analiza,
    'Zeta analiza': Zeta_analiza
})
Z_i_zeta_analiza = \
Z_i_zeta_analiza.apply (lambda x: round (x, 2))
Z_i_zeta_analiza.index = [2021, 2022, 2023]

Z_i_zeta_analiza
```
